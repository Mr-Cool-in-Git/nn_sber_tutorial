```python
import numpy as np
import torch
```

```python
batch_size = 32
input_size = 3
hidden_size = 2
output_size = 1
```

```python
# Create random input and output data
x = torch.randn(128, input_size)
y = torch.randn(128, output_size)
```

```python
nw1 = np.random.random(size=(input_size, hidden_size)).astype('float32')
nw2 = np.random.random(size=(hidden_size, output_size)).astype('float32')

nb1 = np.random.random(size=(y.shape[0], 1)).astype('float32')
nb2 = np.random.random(size=(y.shape[0], 1)).astype('float32')
```

```python
w1 = torch.tensor(nw1)
w2 = torch.tensor(nw2)

b1 = torch.tensor(nb1)
b2 = torch.tensor(nb2) 
```

```python
learning_rate = 1e-6
for t in range(2):    
    
    #TODO
    h1 = x.mm(w1).add(b1).clamp(min=0)
    y_pred = h1.mm(w2).add(b2)
    # Compute and print loss
    loss = (y_pred - y).pow(2).sum().item()
    
    s = 2 * (y_pred - y)
    grad_h2 = torch.mm(s, w2.T)
    grad_h2[h1 < 0] = 0
    
    grad_w2 = torch.mm(h1.T,s)
    grad_b2 = torch.sum(h1).item()

    grad_w1 = torch.mm(x.T, grad_h2)
    grad_b1 = torch.sum(grad_h2).item()
    
    print(t, loss,grad_w1,grad_w2,grad_b1,grad_b2)

    w1 -= learning_rate * grad_w1
    b1 -= learning_rate * grad_b1
    w2 -= learning_rate * grad_w2
    b2 -= learning_rate * grad_b2
```

```python
w1 = torch.tensor(nw1, requires_grad=True)
w2 = torch.tensor(nw2, requires_grad=True)

b1 = torch.tensor(nb1, requires_grad=True)
b2 = torch.tensor(nb2, requires_grad=True) 

import torch.optim as optim

loss_fn = torch.nn.MSELoss(reduction='sum')

learning_rate = 1e-6
optimizer = torch.optim.SGD([w1,w2,b1,b2], lr=learning_rate)

for t in range(1):
    optimizer.zero_grad()
    
    y_pred = x.mm(w1).add(b1).clamp(min=0).mm(w2).add(b2)
    
    loss = loss_fn(y_pred, y)
    
    print(t, loss.item())
    w1.register_hook(lambda grad: print(grad))
    w2.register_hook(lambda grad: print(grad))
    b1.register_hook(lambda grad: print(grad))
    b2.register_hook(lambda grad: print(grad))

    loss.backward()
   
    optimizer.step()
```
